```python
import numpy as np
import matplotlib.pyplot as plt
```

## Hand crafted feature extraction

```python
models = ['Logistic Regression', 'Perceptron', 'Decision Tree', 'Random Forest', 'Gradient Boosting', 
          'Ada Boost', 'XG Boost', 'KNN', 'SVM', 'Cat Boost']
```

```python
accuracy = np.array([99.38, 99.20, 99.25, 99.57, 99.55, 99.38, 99.55, 99.08, 99.40, 99.56])
precision = np.array([99.33, 99.30, 98.62, 99.62, 99.60, 99.30, 99.55, 99.37, 99.37, 99.65])
recall = np.array([98.79, 98.75, 98.99, 99.10, 99.04, 98.88, 99.13, 98.70, 98.70, 99.02])
f1_score = np.array([99.06, 99.02, 98.81, 99.36, 99.32, 99.09, 99.34, 99.03, 99.03, 99.34])
```

```python
plt.figure(figsize = (5, 6))
plt.grid(False)
plt.barh(y=models, width=accuracy)
plt.xlim(88, 100)
plt.ylabel("Machine Learning Models")
plt.xlabel("Accuracy")
plt.show()
```

```python
plt.figure(figsize = (5, 6))
plt.grid(False)
plt.barh(y=models, width=precision)
plt.xlim(88, 100)
plt.ylabel("Machine Learning Models")
plt.xlabel("Precision")
plt.show()
```
